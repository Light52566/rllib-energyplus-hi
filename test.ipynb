```python
from typing import Any, Dict, List, Optional, Tuple, Union

import gymnasium as gym

def post_process_action(action: Union[float, List[float]]) -> Union[float, List[float]]:
    actual_range = (0, 40.0)
    action_space = gym.spaces.Box(low=0, high=40.0, shape=(1,), dtype=float)
    return _rescale(
        n=float(action), range1=(action_space.low, action_space.high), range2=actual_range  # noqa  # noqa
    ) [0]

def _rescale(n: int, range1: Tuple[float, float], range2: Tuple[float, float]) -> float:
    delta1 = range1[1] - range1[0]
    delta2 = range2[1] - range2[0]
    return (delta2 * (n - range1[0]) / delta1) + range2[0]
post = post_process_action(10)
print(post, type(post))
print(isinstance(post, float))
```
